```python
# CELDA 1
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

print("Cargando modelo BETO...")

# Cargar tokenizer y modelo
tokenizer = AutoTokenizer.from_pretrained('../models/beto/tokenizer')
model = AutoModelForSequenceClassification.from_pretrained('../models/beto/modelo')

print("✅ Modelo cargado")

# CELDA 2
def predecir(texto):
    """Predice si un texto es ofensivo o no"""
    
    # Tokenizar
    inputs = tokenizer(texto, return_tensors="pt", padding=True, 
                      truncation=True, max_length=128)
    
    # Predecir
    with torch.no_grad():
        outputs = model(**inputs)
    
    # Obtener probabilidades
    probs = torch.softmax(outputs.logits, dim=1)
    prediccion = torch.argmax(probs, dim=1).item()
    confianza = probs[0][prediccion].item()
    
    etiqueta = "Ofensivo" if prediccion == 1 else "No ofensivo"
    
    return {
        "texto": texto,
        "prediccion": etiqueta,
        "confianza": f"{confianza*100:.2f}%"
    }

# CELDA 3 - Probar
ejemplos = [
    "eres un gilipollas",
    "hola qué tal cómo estás",
    "no eres tonto",
    "eres tonto",
    "vete a la mierda"
]

for ejemplo in ejemplos:
    resultado = predecir(ejemplo)
    print(f"\n{resultado}")
```
