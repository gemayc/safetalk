# 🤖 Generación de Ejemplos Sintéticos 

**Objetivo:** Generar 2,801 ejemplos sintéticos de bullying/hate speech

**Método:** OpenAI API (gpt-4o-mini)

---

```python
import pandas as pd
import numpy as np
from dotenv import load_dotenv
import os
from openai import OpenAI

# Cargar variables del archivo .env
load_dotenv()

# Leer la key
client = OpenAI(api_key=os.environ.get("OPENAI_API_KEY"))

print("✅ Cliente OpenAI configurado")
```

```python

df_limpio = pd.read_csv('../data/processed/zenodo_limpio.csv')

print("="*70)
print("ANALISIS DE EJEMPLOS REALES")
print("="*70)

# Filtrar solo clase 1 (ofensivos)
ejemplos_reales = df_limpio[df_limpio['label'] == 1]['texto'].tolist()

print("\nEjemplos reales de clase 1:", len(ejemplos_reales))

# Analizar caracteristicas
longitudes = [len(texto) for texto in ejemplos_reales]
palabras = [len(texto.split()) for texto in ejemplos_reales]

print("\nCaracteristicas de ejemplos reales:")
print("   Longitud promedio:", int(np.mean(longitudes)), "caracteres")
print("   Palabras promedio:", round(np.mean(palabras), 1), "palabras")
print("   Rango:", min(longitudes), "-", max(longitudes), "caracteres")

# Tomar muestra para mostrar a ChatGPT como referencia
muestra_referencia = np.random.choice(ejemplos_reales, 15, replace=False)

print("\nMuestra de ejemplos reales (para referencia):")
for i, ejemplo in enumerate(muestra_referencia, 1):
    print(f"   {i}. {ejemplo[:100]}...")
```
